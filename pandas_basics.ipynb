# Filtering Rows

`df.loc[]` - uses names\
`df.iloc[]` - index location

```python
test_df = pd.DataFrame([(1, 'Ram'), (2, 'Sita'), (1, 'Hari')], columns=('sn', 'name'))
test_df.set_index('sn', inplace=True)

type(test_df.loc[1]) # Returns DataFrame
type(test_df.loc[2]) # Returns Series

# When locating with index, if there is exactly one row with the index, it returns a Series.
# If there are multiple rows with the given index, it returns a dataframe.
# But in case of iloc, even if there are multiple rows with the given index, it returns a Series with the first matched row.
```

```python
df.loc[:, 'DepTime']
# df['DepTime']
```

```python
df.iloc[5] # Returns series
df.iloc[[5, 6, 7]] # Returns data frame

df.iloc[:,1] # Returns series
df.iloc[:,[1, 2, 3]] # Returns data frame
```

# Rolling Methods

the `.rolling()` method in pandas is a powerful tool for performing calculations over a moving window of fixed size. It's often used in time series analysis but works on any numeric sequence.

`df['column'].rolling(window=N).<aggregation>()`


`window=N`: Number of observations to include in each rolling window.

`<aggregation>`: Any function like `.mean()`, `.sum()`, `.max()`, `.std()`, etc.





```python
new_df = df[['DepTime']]

new_df.assign(
    rolled_DepTime=df['DepTime'].rolling(window=3).mean()
)
```

# Merge data

```python
df_DepDelay = df.groupby(['Airline', 'FlightDate'])[['DepDelay']].mean().reset_index()
df_ArrDelay = df.groupby(['Airline', 'FlightDate'])[['ArrDelay']].mean().reset_index()
```

```python
df_DepDelay.merge(df_ArrDelay)
```

```python
pd.merge(df_DepDelay, df_ArrDelay, on = ['Airline', 'FlightDate'])
```

```python
pd.merge(df_DepDelay, df_ArrDelay, on = 'Airline', suffixes=('_dep','_arr'))
```

```python
# If column names are different

pd.merge(df_DepDelay, df_ArrDelay, left_on='Airline', right_on='Airline')
```
